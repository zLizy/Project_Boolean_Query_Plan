```python
import matplotlib.pyplot as plt
from pandas.plotting import andrews_curves
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
def load(flag):
    filepath = 'repository/run_summary_'+flag+'.csv'
    df = pd.read_csv(filepath,index_col=0)
    print(df.shape)
#     print(df.head())
    return df

df_cost = load('cost')
df_cost
```

```python
def load(flag):
    filepath = 'repository/run_summary_'+flag+'_uniform_mix.csv'
    df = pd.read_csv(filepath,index_col=0)
    print(df.shape)
#     print(df.head())
    return df

df_uni_mix = load('cost')
df_uni_mix
```

```python
_optimizer = df_cost[df_cost['approach']=='optimizer']
_baseline = df_cost[df_cost['approach']=='baseline1']
# plt.plot(df_median.loc[m,'cost'],df_median.loc[m,t],'yo',markersize=15)
mix_optimizer = df_mix[df_mix['approach']=='optimizer']
mix_baseline = df_mix[df_mix['approach']=='baseline1']

pow_optimizer = df_pow[df_pow['approach']=='optimizer']
pow_baseline = df_pow[df_pow['approach']=='baseline1']

uni_mix_optimizer = df_uni_mix[df_uni_mix['approach']=='optimizer']
uni_mix_baseline = df_uni_mix[df_uni_mix['approach']=='baseline1']
```

```python
plt.rcParams['figure.dpi'] = 200
fig, ax = plt.subplots()


ax.scatter(_baseline['cost'], _baseline['accuracy'],label='baseline')
ax.scatter(mix_baseline['cost'], mix_baseline['accuracy'],label='baseline-mix')
ax.scatter(uni_mix_baseline['cost'], uni_mix_baseline['accuracy'],label='baseline-uni-mix')
ax.scatter(pow_baseline['cost'], pow_baseline['accuracy'],label='baseline-power_law')


ax.scatter(_optimizer['cost'],_optimizer['accuracy'],label='optimizer')
ax.scatter(mix_optimizer['cost'],mix_optimizer['accuracy'],label='optimizer-mix')
ax.scatter(uni_mix_optimizer['cost'], uni_mix_optimizer['accuracy'],label='optimizer-uni-mix')
ax.scatter(pow_optimizer['cost'],pow_optimizer['accuracy'],label='optimizer-power_law')

plt.ylabel('Cost')
plt.xlabel('Accuracy')
# plt.yscale('logit')
plt.legend()
```

## Pareto ratio

```python
plt.rcParams['figure.dpi'] = 200
key = [2,4,8,16,32]
r = {2:1,4:2,8:3,16:4,15:4,32:5,28:5,31:5,30:5}

def getX(data):
    return [r[item] for item in data['#predicate']]

x_baseline = getX(_baseline)
x_optimizer = getX(_optimizer)

x_pow_baseline = getX(pow_baseline)
x_pow_optimizer = getX(pow_optimizer)

x_mix_baseline = getX(mix_baseline)
x_mix_optimizer = getX(mix_optimizer)

x_uni_mix_baseline = getX(uni_mix_baseline)
x_uni_mix_optimizer = getX(uni_mix_optimizer)
```

```python
fig, ax = plt.subplots()
# ax.scatter(x_baseline, _baseline['pareto_ratio'],label='baseline')
ax.scatter(x_optimizer,_optimizer['pareto_ratio'],label='optimizer')

ax.scatter(x_mix_optimizer,mix_optimizer['pareto_ratio'],label='optimizer-mix')
ax.scatter(x_uni_mix_optimizer,uni_mix_optimizer['pareto_ratio'],label='optimizer-uni-mix')
# ax.scatter(x_pow_baseline, pow_baseline['pareto_ratio'],label='baseline-pow')
ax.scatter(x_pow_optimizer,pow_optimizer['pareto_ratio'],label='optimizer-pow')



plt.xticks(range(1,6),key)
plt.ylabel('Pareto ratio')
plt.xlabel('#predicate')
plt.legend()
```

```python
fig, ax = plt.subplots()
ax.scatter(x_optimizer,_optimizer['pareto_ratio'],label='optimizer')
ax.scatter(x_mix_optimizer,mix_optimizer['pareto_ratio'],label='optimizer-mix')
ax.scatter(x_uni_mix_optimizer,uni_mix_optimizer['pareto_ratio'],label='optimizer-uni-mix')
ax.scatter(x_pow_optimizer,pow_optimizer['pareto_ratio'],label='optimizer-pow')

# rectangular box plot
bplot1 = ax.boxplot([_optimizer[_optimizer['#predicate']==k]['pareto_ratio'] for k in [2,4,8,16,32]],
                     showfliers=False,
                     vert=True,  # vertical box alignment
                     labels=range(5)) 
c1 = "blue"
for item in ['boxes', 'whiskers', 'fliers', 'medians', 'caps']:
    plt.setp(bplot1[item], color=c1)

bplot1 = ax.boxplot([mix_optimizer[mix_optimizer['#predicate']==k]['pareto_ratio'] for k in [2,4,8,16,32]],
                     showfliers=False,
                     vert=True,  # vertical box alignment
                     labels=range(5)) 
c2 = "orange"
for item in ['boxes', 'whiskers', 'fliers', 'medians', 'caps']:
    plt.setp(bplot1[item], color=c2)
    
# bplot1 = ax.boxplot([x_uni_mix_optimizer[x_uni_mix_optimizer['#predicate']==k]['pareto_ratio'] for k in [2,4,8,16,32]],
#                      showfliers=False,
#                      vert=True,  # vertical box alignment
#                      labels=range(5)) 

# c2 = "orange"
# for item in ['boxes', 'whiskers', 'fliers', 'medians', 'caps']:
#     plt.setp(bplot1[item], color=c2)
    
plt.xticks(range(1,6),key)
plt.ylabel('Pareto ratio')
plt.xlabel('#predicate')
plt.legend()
```

```python
def boxplot(data,query_range):
    box_colors = ['darkkhaki', 'royalblue']
    fig, ax = plt.subplots()
    ax.set_title('Pareto Ratio')
    bp = ax.boxplot(data)
    plt.xticks(range(1,len(query_range)+1), query_range)
    plt.xlabel("#queries")
    plt.ylabel("Pareto ratio")
    plt.show()
```
